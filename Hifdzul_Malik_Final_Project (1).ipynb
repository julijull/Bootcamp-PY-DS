```python
selected_features = ['GDPperkapita', 'Pendapatan']
```

```python
# Visualisasi boxplot untuk deteksi outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[selected_features])
plt.title('Boxplot untuk Deteksi Outliers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Deteksi dan penanganan outliers menggunakan metode IQR
def handle_outliers(data, features):
    for feature in features:
        Q1 = data[feature].quantile(0.25)
        Q3 = data[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Menangani outliers
        data[feature] = data[feature].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

    return data

# Deteksi outliers
outliers_exist = False
for feature in selected_features:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    if not outliers.empty:
        outliers_exist = True
        print(f"Outliers ditemukan pada feature '{feature}':")
        print(outliers)

# Penanganan outliers
if outliers_exist:
    print("\nPenanganan outliers:")
    data = handle_outliers(data, selected_features)
    print("Outliers telah ditangani.")
else:
    print("Tidak perlu penanganan outliers.")

# Visualisasi boxplot setelah penanganan outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[selected_features])
plt.title('Boxplot setelah Penanganan Outliers')
plt.xticks(rotation=45)
plt.show()
```

Outliers tidak hilangkan karena ada kemungkinan bahwa outliers tersebut merupakan representasi dari kondisi negara yang pada list dan dengan jika menghilangkannya bisa saja mengubah data, dan terjadi kesalahan saat memutuskan negara yang benar-benar membutuhkan bantuan

# 3. Clustering

1. Scale the Data
2. Decide the number of clusters, you're free to choose the method:
- Elbow method
- Silhouette score method
- Directly decide the number (Elaborate the number and why)

3. Do clustering with the decided amount of cluster.

4. Create the clustering result graph.

```python
scaler = StandardScaler()

# Scaling data
scaled_data = scaler.fit_transform(data[selected_features])
```

```python
inertia = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plotting inertia untuk menentukan jumlah klaster yang optimal (elbow point)
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Melakukan k-means clustering untuk berbagai jumlah klaster
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting silhouette scores untuk menentukan jumlah klaster yang optimal
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()
```

```python
# Melakukan k-means clustering dengan jumlah klaster yang telah diputuskan
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)
```

```python
# Visualisasi hasil clustering
plt.figure(figsize=(8, 6))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=cluster_labels, palette='viridis')
plt.title('Clustering Result')
plt.xlabel('GDPperkapita')
plt.ylabel('Pendapatan')
plt.legend(title='Cluster')
plt.show()
```

```python
# Memilih klaster yang akan difokuskan
focus_cluster_label = 0

# Memilih data untuk klaster yang difokuskan
focus_cluster_data = data[cluster_labels == focus_cluster_label]
```

```python
# Menampilkan negara-negara yang termasuk dalam klaster yang difokuskan
print("Negara-negara yang termasuk dalam klaster yang difokuskan:")
for index in focus_cluster_data.index:
    country_name = data.loc[index, 'Negara']
    print(country_name)
```

```python
# Memilih negara-negara terbaik untuk menerima bantuan
best_countries = focus_cluster_data.nsmallest(10, 'Pendapatan')[['Negara', 'GDPperkapita','Pendapatan',]]
print("Negara-negara terbaik untuk menerima bantuan:")
print(best_countries)
```
